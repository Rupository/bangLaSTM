```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_parquet('bn_data.parquet', columns=['question', 'answer', 'title'])

df
```

```python
df = df.drop_duplicates(subset='title')

df
```

```python
latin_regex = r'[a-zA-Z]'

mask = (df['question'].str.contains(latin_regex, regex=True, na=False)) | \
       (df['answer'].str.contains(latin_regex, regex=True, na=False))

df = df[~mask]

df
```

```python
df = df.drop(columns=['title'])
df = df.reset_index(drop=True)

df
```

```python
df_train, df_temp = train_test_split(df, test_size=0.40, random_state=42)
df_val, df_test = train_test_split(df_temp, test_size=0.5, random_state=42)
```
